```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.cvtColor(cv2.imread("left.jpg"),cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread("right.jpg"),cv2.COLOR_BGR2RGB)
img1 = cv2.UMat(img1)
img2 = cv2.UMat(img2)
```

```python
plt.imshow(cv2.UMat.get(img1))
```

```python
orb = cv2.cuda_ORB.create(
    nfeatures=1000,
    scaleFactor=1.2,
    nlevels=8
)
img1_gpu =  cv2.cuda_GpuMat()
img1_gpu.upload(img1)
print(type(img1_gpu))
kp1, des1 = orb.detectAndCompute(img1_gpu, None)
```
